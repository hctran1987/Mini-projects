## Create target variable (adopted users)

```python
df1['time_stamp'] = pd.to_datetime(df1['time_stamp'])
```

```python
df1.info()
```

```python
list_id = df1.user_id.unique()
```

```python
len(list_id)
```

```python
# Create dictionary with keys as User IDs and values as O or 1 
#(not adopted or adopted).
adopted_user = dict()
for i in list_id:
    a = np.array(df1[df1.user_id == i]['time_stamp'].sort_values())
    if len(a) < 3:
       adopted_user[i] = 0
    else:
        for j in range(len(a)-2):
            if a[j+2] < a[j] + pd.DateOffset(days=7):
                adopted_user[i] = 1
                break
            else:
                adopted_user[i] = 0
                
            
    
```

```python
adopted_user
```

```python
# Transform dictionary of adopted users into dataframe 
df_adopted_user = pd.DataFrame.from_dict(adopted_user, orient='index', columns = ['adopted_user']).reset_index()
```

```python
df_adopted_user
```

## Form the data for the analysis, clean them, and extract features for analysis

```python
# Combine predictor variables and target variable 
data = df_adopted_user.merge(df, left_on = 'index', right_on = 'object_id')
```

```python
data
```

```python
# Extract the user email domain
import re as re
def email_domain(text):
    email = re.findall(r'@[\w\.-]+',str(text))
    return email[0][1:-4] #",".join(email)
```

```python
data['email_domain'] = data['email'].astype('str').apply(lambda x: email_domain(x))
```

```python
data['email_domain'].unique()
```

```python
def email(x):
    if x in list_email:
        return x
    else:
        return 'other'
        
```

```python
data['email_domain'] = data['email_domain'].apply(lambda x: email(x))
```

```python
# Create the new predictor variable (number_of_using_days)
from datetime import datetime
```

```python
data['last_session_creation_time'] = data['last_session_creation_time'].astype('int').astype("datetime64[s]")
data['creation_time'] = pd.to_datetime(data['creation_time'])
```

```python
data['number_of_using_days'] = (data['last_session_creation_time'] - data['creation_time']).dt.days
```

```python
# Extract features for analysis
data_for_use = data[['creation_source', 'opted_in_to_mailing_list',
                    'enabled_for_marketing_drip', 'email_domain', 'number_of_using_days']]
y = data['adopted_user']
```

## Choose model for analysis

```python
# I check the ratio between 0 values and 1 values to choose the metric for the analysis later
data['adopted_user'].value_counts(normalize = True)
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(data_for_use_numerized, y, test_size = 0.2, 
                                                    stratify = y, random_state = 42)
```

```python
# The target variable data is skewed I decide to use ROC_AUC score as the metric to choose the model. Also, I am
# looking for the factors to predict the future user adoption. So I make a priority on the models that can help me
# extract the feature important easily.
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import roc_auc_score, classification_report
from xgboost import XGBClassifier
```

```python
dt = DecisionTreeClassifier(random_state = 42)
rf = RandomForestClassifier(random_state = 42)
ada = AdaBoostClassifier(random_state = 42)
GB = GradientBoostingClassifier(random_state = 42)
XGB = XGBClassifier(random_state = 42)
```

```python
np.mean(cross_val_score(dt, X_train, y_train, cv=5, scoring = 'roc_auc'))
```

```python
np.mean(cross_val_score(rf, X_train, y_train, cv=5, scoring = 'roc_auc'))
```

```python
np.mean(cross_val_score(ada, X_train, y_train, cv=5, scoring = 'roc_auc'))
```

```python
np.mean(cross_val_score(GB, X_train, y_train, cv=5, scoring = 'roc_auc'))
```

```python
np.mean(cross_val_score(XGB, X_train, y_train, cv=5, scoring = 'roc_auc'))
```

```python
# We proceed with the Ada Boost and try to optimize the number of estimators.
from sklearn.model_selection import GridSearchCV
```

```python
parameters = {'n_estimators':[1, 2, 5, 10, 20, 50, 100, 200, 500, 1000]}
grid = GridSearchCV(ada, parameters, cv = 5)
```

```python
grid.fit(X_train, y_train)
```

```python
best_model = grid.best_estimator_
```

```python
best_model
```

```python
# We fit the best model to the whole train data and see its performance on the test data
best_model.fit(X_train, y_train)
```

```python
y_test_pred = best_model.predict(X_test)
```

```python
print('The AUC score of the best model on the test set:', roc_auc_score(y_test, y_test_pred))
```

```python
print('The classification report of the best model on the test set:\n', classification_report(y_test, y_test_pred))
```

## Discuss the factors that predict future adopted users

```python
#Since the chosen model performed so well on the data we can choose it to see the important factors for the prediction
important_features = pd.DataFrame(best_model.feature_importances_, index = data_for_use_numerized.columns, 
                                  columns = ['Contributions'])
```
